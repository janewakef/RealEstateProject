```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
source = requests.get('https://www.cbssports.com/nfl/draft/news/undrafted-free-agent-tracker-2020-rookie-signings-from-all-32-nfl-teams-in-one-spot-following-nfl-draft/').text
```

```python
colNames = ['Division','Team','Player Info']
undraftedPlayers = pd.DataFrame(columns=colNames)
```

```python
soup = BeautifulSoup(source, 'lxml')
main = soup.find('div', {"id": "article-main-body"})
teams = main.find('div', class_ = 'row')
```

```python
for header in teams.find_all('h2'):
    # division name 
    divisionNameString = header.text
    for elem in header.next_siblings:
        if (elem.name == 'p'):
            teamName = elem
            # team name 
            teamNameString = teamName.text
            playerListUL = elem.find_next_sibling('ul')
            for player in playerListUL.find_all('li'):
                # player name 
                playerInfoString = player.text
                # this is where I add all the vriables into the data frame 
                playerList = [divisionNameString, teamNameString, playerInfoString]
                addOn = dict(zip(undraftedPlayers.columns, playerList))
                undraftedPlayers = undraftedPlayers.append(addOn, ignore_index=True)
        if (elem.name == 'h2'):
            break
```

```python
undraftedPlayers.head(5)
```

```python
#undraftedPlayers['Name2'] = undraftedPlayers['Player Info'].str.split(",",expand=True)[0]
# Best way to do it:
undraftedPlayers[['Name','Position', 'School']] = undraftedPlayers['Player Info'].str.split(",",expand=True)
undraftedPlayers['Name'] = undraftedPlayers['Name'].str.strip()
undraftedPlayers['Position'] = undraftedPlayers['Position'].str.strip()
undraftedPlayers['School'] = undraftedPlayers['School'].str.strip()
```

```python
undraftedPlayers.drop('Player Info', axis = 1, inplace = True)
```
